### Il seguente Notebook permette di correggere il path presente nel file .json delle annotazioni delle immagini automaticamente labellate 

```python
import os
import json
import random
import string
```

Inserire nella chiave 'info['images_dir']' il path corretto in cui sono contenute le immagini warp.

```python
info = {
    'annotation_dir': "../../dataset/elettrocablaggi_20200921/train/annots/",
    'images_dir': "C:/Users/lsnqr/Documents/DLVSystem/Map4ID/code/Mask_RCNN/dataset/elettrocablaggi_20200921/train/warp_data/" 
       }
```

```python
def save(new_file_name, old_annot, images_path, annots_path):
    new_annot = dict()
    new_annot['regions'] = list()

    width = old_annot['asset']['size']['width']
    height = old_annot['asset']['size']['height']
    for region in old_annot['regions']:
        new_annot['regions'].append(region)

    new_annot['asset'] = {"format": "JPG",
                          "id": ''.join(
                              random.choice(string.ascii_uppercase + string.digits) for _ in range(10)),
                          "name": new_file_name[:-5] + '.jpg',
                          "path": images_path + new_file_name[:-5] + '.jpg',
                          "size": {
                              "width": width,
                              "height": height
                          },
                          "state": 2,
                          "type": 1
                          }
    with open(os.path.join(annots_path, new_file_name), 'w') as outfile:
        json.dump(new_annot, outfile, indent=4)
```

```python
annotations = [f for f in os.listdir(info['annotation_dir']) if os.path.isfile(os.path.join(info['annotation_dir'], f))]

for a in annotations:
    f = open(os.path.join(info['annotation_dir'], a), encoding='utf-8')
    if a not 'labels.txt':
        annotation = json.load(f)
        save(a, annotation, info['images_dir'], info['annotation_dir'])
    f.close()

```
